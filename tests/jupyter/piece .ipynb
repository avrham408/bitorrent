```python
pieces = []
counter = 0
for i in range(0, len(torrent_file.pieces), 20):
    counter += 1
    if i != len(torrent_file.pieces) - 20:
        piece = Piece(torrent_file.pieces[i:i+20] , torrent_file.piece_length , counter)
    else:
        piece = Piece(torrent_file.pieces[i:i+20] , torrent_file.length % torrent_file.piece_length, counter)
    pieces.append(piece)
```
